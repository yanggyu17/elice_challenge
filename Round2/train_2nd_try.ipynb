# Data

```python
total_h, total_a = [], []

for i in range(len(data)):
    for j in range(10):
        if data[i]['participants'][j]['teamId'] == 100:
            total_h.append(data[i]['participants'][j]['championId'])
            total_h.append(data[i]['participants'][j]['summonerId'])
            #total.append(data[i]['participants'][j]['teamId'])
        else:
            total_a.append(data[i]['participants'][j]['championId'])
            total_a.append(data[i]['participants'][j]['summonerId'])
            #total.append(data[i]['participants'][j]['teamId'])
```

```python
total_h = np.reshape(total_h, (38679, 10))
total_a = np.reshape(total_a, (38679, 10))
total_h
```

```python
df_total_h = pd.DataFrame(total_h, columns=[['participants','participants','participants','participants','participants','participants','participants','participants','participants','participants'],
                                          ['team(100)','team(100)','team(100)','team(100)','team(100)','team(100)','team(100)','team(100)','team(100)','team(100)'],
                                          ['championId','summonerId','championId','summonerId','championId','summonerId','championId','summonerId','championId','summonerId']])
df_total_h.head()
```

```python
df_total_a = pd.DataFrame(total_a, columns=[['participants','participants','participants','participants','participants','participants','participants','participants','participants','participants'],
                                          ['team(200)','team(200)','team(200)','team(200)','team(200)','team(200)','team(200)','team(200)','team(200)','team(200)'],
                                          ['championId','summonerId','championId','summonerId','championId','summonerId','championId','summonerId','championId','summonerId']])
df_total_a.head()
```

```python
df_data = pd.concat([df_total_h, df_total_a], axis=1)
df_data.head()
```

```python
df_all = pd.concat([df_data, df_w], axis=1)
df_all.tail()
```

```python
# split data into X and y
dfX = df_all.iloc[:,:-1]
dfy = df_all.iloc[:,[-1]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dfX, dfy, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
#from xgboost import XGBClassifier

rf_clf = RandomForestClassifier(n_estimators=200, max_depth=10)
#xgb_clf =  XGBClassifier(n_estimators=300, max_depth = 7, learning_rate=0.1 )
```

```python
rf_clf.fit(X_train, y_train)
#xgb_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

target = ['100','200']

y_pred = rf_clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=target))

# for clf in (rf_clf):
#     clf.fit(X_train, y_train)
#     y_pred = clf.predict(X_test)
#     print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
#     print (classification_report(y_test, y_pred, target_names=target))
```
