```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
vect.fit(only_text)
```

```python
bag_of_words = vect.transform(only_text)
print("bag_of_words:\n{}".format(repr(bag_of_words)))
```

```python
feature_names = vect.get_feature_names()
print('특성 개수: {}'.format(len(feature_names)))
print('처음 20개 특성: {}'.format(feature_names[:20]))
print('10000부터 10020까지 특성: {}'.format(feature_names[10000:10020]))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

scores = cross_val_score(LogisticRegression(), bag_of_words, only_y, cv=5 )
print('교차 검증 평균 점수 : {:.2f}'.format(np.mean(scores)))
```
